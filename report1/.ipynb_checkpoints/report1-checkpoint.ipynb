#### Stacking Faults
Il difetto su cui mi sono concentrato fin da subito è lo stacking-fault(SF): infatti la descrizione dello SF come sfasamento nell'alternanza ABC dei piani cristallini dell'FCC si riferisce ai piani normali alla direzione $[111]$. Per questo ho primariamente ruotato la cella primitiva dell'FCC in modo da avere $z'=[111]$, e completato la nuova base con le scelte $x'=[10-1]$ e $y'=[-12-1]$ (la vecchia era la base "canonica" $x=[100]$ ...). A questo punto la successione ABC di piani cristallini si verifica lungo l'asse $z$, e quindi lo stacking fault si ottiene "traslando" in maniera opportuna un piano $xy$. 

La cella primitiva in questo caso contiene 6 atomi, e quindi abbiamo deciso con Francesco di allontanarci dalla cella cubica con l'intento di mantenere il numero di atomi contenuti nella cella il più vicino possibile a 2048, la scelta di Andrea. La cella utilizzata ha quindi queste specifiche:

| |dimensioni($\mathring A$) | # partic. (intera)    | # partic. (cilindrata)| # particelle per piano| numero di piani| 
|-|---------- |-----------------------|-----------------------|----------------|-----------------------|
| Cella corrente|45.79082094 x 39.65601419 x 46.73506092 |2160|2060|144 | 15 |
| Cella di Andrea| 43.172 x 43.172 x 43.172 |  2048 | 2030 | 128 | 16 |


Tenendo conto che ho tagliato la cella lasciando invariato il protocollo di Andrea, che consisteva nel selezionare una porzione cilindrica con area di base uguale alla faccia del cubo (nel mio caso faccia xy) e di altezza uguale all'altezza del cubo(z nel mio caso). Non mi aspetterei che *nel caso ideale* questi cambiamenti influenzino il fattore di struttura nè la CCA, soprattutto perchè la rete neurale poi vede soltanto le medie di queste quantità, però non saprei immaginare se questo possa cambiare con l'inserimento dei difetti.

    con Francesco ci siamo chiesti se sia necessario o meno che il taglio cilindrico contenga i difetti. Questo succede sicuramente per costruzione per gli SF, ma non è detto che succeda per lacune, atomi interstiziali, dislocazioni. Sono convinto che in ogni caso S(q) e la CCA contengano informazione relativa alla presenza di difetti, perchè il sistema verrà tagliato dopo un'evoluzione temporale (che in questo caso comprende anche la transizione di fase), e quindi mi aspetto che il risultato finale sia influenzato dai difetti. Ciononostante, non saprei "quantificare" questa informazione nè quindi saprei cosa aspettarmi dal far vedere alla rete questo dato. La domanda mi sembra interessante in sè, ma all'atto pratico si scarta circa il 5% degli atomi con l'operazione di taglio, e comunque ogni dump immagino sarà ruotato 100 volte seguendo il protocollo di Andrea, e quindi i casi in cui questo si verifica diventano probabilmente marginali.

Qualche osservazione(continua attraverso le figure):
-  $ABCABCABC\to ABABCABCA$ è il mapping tra la vecchia e nuova sequenza di stacking. Siccome ci sono 15 piani e lo stacking dei piani è periodico modulo 3, anche nella nostra cella con le pbc ci troviamo lo stacking di due piani AA
- la situazione è analoga nel caso in cui si sostituisse un piano con un altro lasciando intatta la sequenza superiore; solo, la situazione "frustrata" AA non sarebbe alle pbc

- l'evoluzione temporale del sistema attraverso LAMMPS porta lo stacking AA ad formare a volte un FCC, a volte un HCP, almeno secondo la common neighbour analysis(CNA) di ovito, con tendenza a formare FCC quando lo SF inserito si trova vicino ai bordi della cella

- l'energia totale media del cristallo nel caso in cui lo stacking fault sia "lontano dal bordo della cella" non è compatibile con quella di alcuni dei casi 09z e 01z.

- la nomenclatura è un po' fuorviante: essendo 15 i piani  ogni incremento di 0.1 mette lo SF in media 1.5 piani spostato rispetto al precedente; questo implica che i casi 0.1z e 0.9z non sono simmetrici per inversione dell'asse z:

```python
plt.figure(figsize=(18,9))
plt.subplot(121,frameon=False)
plt.title('0.1z, rosso: HCP; verde: FCC; bianco: Other')
plt.imshow(img.imread('start_mixture01.png'))
plt.axis('off')
plt.subplot(122,frameon=False)
plt.title('0.9z, rosso: HCP; verde: FCC; bianco: Other')
plt.imshow(img.imread('start_mixture09.png'))
plt.axis('off');
```

- si vede bianco nella common neighbour analysis perchè in pbc sono due piani AA quindi non appartengono ne a FCC ne a HCP

- il sistema 01z fa uno stacking fault in più! Gia dopo 3000-4000 timesteps, si forma un nuovo SF a metà cristallo, e quindi il sistema si trova con 3 SF presenti. Questo è visto anche dai local bonds nel piano Q4-Q6

```python
plt.figure(figsize=(18,9))
plt.subplot(141,frameon=False)
plt.title('Timestep=3000, rosso: HCP; verde: FCC; bianco: Other')
plt.imshow(img.imread('01s0dump3.png'))
plt.axis('off')
plt.subplot(142,frameon=False)
plt.title('Timestep=4000')
plt.imshow(img.imread('01s0dump4.png'))
plt.axis('off')
plt.subplot(143,frameon=False)
plt.title('Timestep=5000')
plt.imshow(img.imread('01s0dump5.png'))
plt.axis('off')
plt.subplot(144,frameon=False)
plt.title('Timestep=6000')
plt.imshow(img.imread('01s0dump6.png'))
plt.axis('off');
```
